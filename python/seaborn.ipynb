```python
#单特征快速绘图
#seaborn 提供了distplot方法，默认绘制的是直方图，并且带有KDE核密度估计函数
data = np.random.normal(size=100)
sb.set()
plt.figure(figsize=(15,8))
sb.set_context("poster",font_scale=0.9,rc={"lines.linwidth":2.5})
sb.distplot(data)
```

```python
#直方图的参数设置，弃用kde核密度函数，增添“小地毯”设置rug关键字参数为True
plt.figure(figsize=(15,8))
sb.distplot(data,kde=False,rug=True)
```

```python
#直方图是对连续数据划分bins进行表示的，bins的个数可以通过bins关键字参数指定
plt.figure(figsize=(15,8))
sb.distplot(data,kde=False,rug=True,bins=20)
```

```python
#只绘制kde
plt.figure(figsize=(15,8))
sb.distplot(data,kde=True,rug=True,hist=False)
```

```python
#只绘制kde
plt.figure(figsize=(15,8))
sb.kdeplot(data,shade=10)
```

```python

#通过设置bw bandwidth参数指定kden拟合的精度，类似于直方图的bins效果
#只绘制kde,bw越小精度越高
plt.figure(figsize=(15,8))
sb.kdeplot(data)
sb.kdeplot(data,bw=.1,label="bw:0.1")
sb.kdeplot(data,bw=3, label ="bw:3")
plt.legend()
sb.distplot(data,kde=True,rug=True,hist=False)
```

```python
plt.figure(figsize=(15,8))
sb.kdeplot(data,shade=True,cut=-1)
sb.rugplot(data)
#设置cut的不同部分就能截取得到
```

```python
plt.figure(figsize=(15,8))
x=np.random.gamma(6,size=100)
sb.distplot(x,kde=10,fit=stats.gamma)
#使用参数估计直观看出整个数据集的分布情况，和kde核密度函数类似
#dstplot是直方图
```

```python
#二元数据的分布图绘制
#使用numpy中mutivariate_normal方法产生多元正太分布的数据，第一个参数为每个维度的均值，第二个参数为多个维度的协方差
mean,cov =[0,1],[(1,.5),(.5,1)]
data = np.random.multivariate_normal(mean,cov,200)
df= pd.DataFrame(data,columns=["x","y"])
df.head()
```

```python
#mattplotlib中最常见就是绘制散点图，dataFRame是可以直接绘制模块的
sb.set_context("notebook")
df.plot.scatter("x","y")
```

```python
plt.scatter(df["x"],df["y"])
```

```python
#使用seaborn绘中的joinplot绘制散点图
sb.jointplot(x="x",y="y",data=df)
```

```python
#hexbinplot蜂窝图
#joinplot中增加kind关键字参数，指定hex,color用于指定的颜色
x,y = np.random.multivariate_normal(mean,cov,1000).T
with sb.axes_style("whitegrid"):
    sb.jointplot(x=x,y=y,kind="hex",color="g")
```

```python
#joinplot中通过kind参数指定"kde"参数，绘制核密度曲线
x,y = np.random.multivariate_normal(mean,cov,1000).T
with sb.axes_style("whitegrid"):
    sb.jointplot(x=x,y=y,kind="kde",color="g")
```

```python
#seaborn提供了kdeplot方法用于绘制密度曲线，但是常用到matplotlib中的axes
#对象，但是matplotlib是自己管理这个对象
f,ax = plt.subplots(figsize=(6,6))
print(type(f),type(ax))
#kdeplot方法接受x,y和ax
sb.kdeplot(df.x,df.y,ax=ax)
#指定vertical垂直方向
sb.rugplot(df.y,color="g", ax=ax)
sb.rugplot(df.y,vertical = True, ax=ax)
```

```python
f,ax = plt.subplots(figsize=(6,6))
sb.rugplot(df.y,vertical = True, ax=ax)
cmap=sb.cubehelix_palette(as_cmap=True,dark=0.,light=1,reverse=True)
print(cmap.colors)
sb.kdeplot(df.x,df.y,cmap=cmap,n_levels=60,shade=True)
#通过关键字指定cmap,生成颜色面板，返回colormap
```

```python
##joinGrid，joinplot会返回这个对象，可以通过这个对象对图表进行更多的设置
g=sb.jointplot(x="x",y="y",data=df,kind="kde",color="y")
print()
g.plot_joint(plt.scatter,c="r",s=20,linewidth=1,marker=".")
g.ax_joint.collections[0].set_alpha(0)
#设置轴坐标标签,s=size,c=color,marker="."
g.set_axis_labels("$X$","$Y$")
#一种latex的数学表达式
```

```python
plt.figure(figsize=(10,8))

#使用boxplot方法绘制盒图，盒图包含最大值，最小值，中位数，上四分位数Q3
#下四分位数Q1,及离群点信息，IQR=Q3-Q1也是盒子的长度，上下的横线是
#最大值和最小值，超过1.51QR的数据单独绘制，被认为是离群点
#hue用于指定按某个类别绘制不同的颜色，列入指定hue="time"
sb.boxplot(x="day",y="total_bill",data=tips,hue="time")
```

```python
tips["weekend"]=tips["day"].isin(["Sat","Sun"])
sb.boxplot(x="day",y="total_bill",data=tips,dodge=True,hue="weekend")
```

```python
#使用小提琴图，更加直观的描述类别数据内部的分布情况
#小提琴图的实现中，用到了一个KDE核密度估计函数，用于更好的描述数据的分布
#可以看到小提琴图包含了盒图的所有能描绘的信息，而且二外增加了核密度估计函数
#小提琴图的离群点也是在绘制图中，而不像盒图那样单独的绘制
plt.figure(figsize=(10,8))
sb.violinplot(x="total_bill",y="day",hue="time",data=tips,dodge=True)
```

```python
#小提琴图是包含所有的信息，白色的点是中位数，上端是上四分位数，下端是下四分位数
#黑色指针最尖端的点为最大值，整个轮廓是将异常值揽括在列，我们可以
#通过bw这个，参数用来平衡指定密度函数的他的带宽
sb.violinplot(bw=0.2,x="total_bill",y="day",hue="time",data=tips,dodge=True)
```

```python
#即使翻看文档，对小提琴图进行分割，当然这种情况适合两种类别的数据，因为中轴线只有两边
#通过hue关键字参数指定分割的数据，并且指定split为True

sb.violinplot(x="day",y="total_bill",hue="sex",data=tips,split=True,bw=0.2)
```

```python
#通过inner关键字参数，显示每个内部的数据详细信息，而不是上面的汇总信息
sb.violinplot(x="day",y="total_bill",hue="sex",data=tips,palette="coolwarm",inner="stick",split=True,bw=0.2)
```

```python
#violinplot和swarmplot的组合使用
plt.figure(figsize=(10,8))
sb.violinplot(x="day",y="total_bill",data=tips,innner=None)
sb.swarmplot(x="day",y="total_bill",data=tips,color="w",alpha=.6)
```
